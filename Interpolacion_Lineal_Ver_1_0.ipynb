```python
import numpy as np
import matplotlib.pyplot as plt

def interpolacion_lineal_vectores(x_vals, y_vals, x):
    """
    Función para realizar la interpolación lineal a partir de vectores de datos.

    Parámetros:
    x_vals: Lista o arreglo de los valores de x conocidos.
    y_vals: Lista o arreglo de los valores de y conocidos.
    x: Valor de x donde se quiere estimar y.

    Retorna:
    El valor interpolado de y para el valor dado de x.
    """
    # Verificar que los vectores de x y y tienen el mismo tamaño
    if len(x_vals) != len(y_vals):
        raise ValueError("Los vectores de x y y deben tener la misma longitud.")

    # Verificar que el valor de x esté dentro del rango de x_vals
    if x < min(x_vals) or x > max(x_vals):
        raise ValueError("El valor de x está fuera del rango de los datos.")

    # Encontrar los dos puntos entre los que se encuentra x
    for i in range(len(x_vals) - 1):
        if x_vals[i] <= x <= x_vals[i + 1]:
            # Aplicar la fórmula de interpolación lineal entre los puntos (x_vals[i], y_vals[i]) y (x_vals[i+1], y_vals[i+1])
            y = y_vals[i] + (y_vals[i + 1] - y_vals[i]) / (x_vals[i + 1] - x_vals[i]) * (x - x_vals[i])
            return y

# Ejemplo de uso:
# Vectores de datos conocidos
x_vals = np.array([0, 10, 20, 30, 40])
y_vals = np.array([0, 85, 60, 45, 30])

# Valor de x en el cual queremos interpolar
x = 15

# Llamamos a la función de interpolación
y_interpolado = interpolacion_lineal_vectores(x_vals, y_vals, x)

# Graficar los datos y la interpolación
plt.figure(figsize=(8, 6))

# Graficamos los puntos originales
plt.plot(x_vals, y_vals, 'o-', label='Datos conocidos')

# Graficamos el punto interpolado
plt.plot(x, y_interpolado, 'ro', label=f'Interpolado en x={x}, y={y_interpolado:.2f}')

# Añadimos etiquetas y título
plt.title('Interpolación Lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.show()

# Imprimir el resultado
print(f"El valor interpolado de y para x = {x} es: {y_interpolado:.2f}")
```
